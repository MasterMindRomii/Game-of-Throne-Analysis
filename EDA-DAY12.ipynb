# Import Neccessary Libraries

# Read the Csv Files

```python
import glob

# List all CSV files in a directory
csv_files = glob.glob(r'C:\Users\Dell\Desktop\75 Day\EDA-DAY12/*.csv')

# Initialize an empty list to store DataFrames
dfs = []

# Read each CSV file and append it to the list
for file in csv_files:
    df = pd.read_csv(file)
    dfs.append(df)

# Concatenate all DataFrames into one
combined_df = pd.concat(dfs, ignore_index=True)

# Now, combined_df contains all the data from the multiple CSV files
```

# Data Visualization

```python
sns.set_style('darkgrid')
ax=sns.countplot(x='Type',data=df,hue='weight')
# plt.figure(figsize=(12,8))
plt.xlabel('Type')
plt.ylabel('weight')
plt.show()
```

```python
plt.figure(figsize=(48,34))
sns.histplot(data=df,x='Source', bins=10,kde=True)
plt.title("Source Vs. Target")
plt.xlabel("Source")
plt.ylabel("Target")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="weight",data=df,hue="Target")
plt.title("Book Vs. Target")
plt.xlabel("weight")
plt.ylabel("Target")
plt.xticks(rotation=90)
plt.show()
```

# Refrence Visualization

```python
# Data Visualization 1: Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Source', y='weight')
plt.title('Weight of Source Characters')
plt.xticks(rotation=90)
plt.xlabel('Source')
plt.ylabel('Weight')
plt.show()
```

```python
# Data Visualization 2: Scatter Plot
plt.figure(figsize=(30, 6))
sns.scatterplot(data=df, x='Source', y='weight', size='weight', sizes=(50, 200))
plt.title('Weight vs. Source with Target Information')
plt.xticks(rotation=90)
plt.xlabel('Source')
plt.ylabel('Weight')
plt.show()
```

```python
# Data Visualization 3: Histogram
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='weight', bins=5, kde=True)
plt.title('Weight Distribution')
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.show()
```

```python
# Data Visualization 4: Count Plot
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Type')
plt.title('Count of Types')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```
